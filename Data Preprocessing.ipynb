```python
#Imports
import pandas as pd
```

```python
#Read in CSV file
crime_df = pd.read_csv('/content/Major_Crime_Indicators_Open_Data.csv', index_col=None, header=0)


#Drop empty rows
crime_df.isnull().sum()

#Want all rows where location is related to TTC
relevant_locations = ['Ttc Subway Station', 'Ttc Subway Train', 'Ttc Street Car', 'Ttc Bus', 'Ttc Bus Stop / Shelter / Loop', ]


#Filter out rows not related to TTC
crime_df = crime_df[crime_df['LOCATION_TYPE'].isin(relevant_locations)]

#Resulting dataframe size:
print("The resulting df is: ", crime_df.shape)

#Drop all columns except for OCC_DATE, NEIGHBOURHOOD_158, MCI_CATEGORY.
crime_df = crime_df[['OCC_DATE','NEIGHBOURHOOD_158', 'MCI_CATEGORY']]

#Changing date format
crime_df['OCC_DATE'] = pd.to_datetime(crime_df['OCC_DATE']).dt.strftime('%Y/%m/%d')


print(crime_df.head)
print(crime_df.columns)
#Write to a new CSV file
#crime_df.to_csv('processed_crime_dataset.csv')
```
